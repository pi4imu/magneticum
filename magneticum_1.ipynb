```python
#%matplotlib notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d  
%matplotlib inline

#!ls
```

AGN

```python
data_name = "../eROSITA_cluster_agn_box2hr/agn_143.dat"
headers1 = ["ID", "x [kpc/h]", "y [kpc/h]", "z[kpc/h]", "M [Msol]", "Mdot [Msol/year]", "Lbol [erg/s]", "Lsoftx [erg/s]", "Lhard [erg/s]", "f_obs"]
table1 = pd.read_csv(data_name, comment='#', header=None, sep='\s+', names=headers1)#, names=headers)
table1
```

```python
plt.rcParams["figure.figsize"] = [14.00, 7.00]
#plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

data_table = table1.to_numpy().astype(float)

x = data_table[:,1]
y = data_table[:,2]
z = data_table[:,3]
c = data_table[:,9]

#ax.set_xlabel('X Label')
#ax.set_ylabel('Y Label')
#ax.set_zlabel('Z Label')

img = ax.scatter(x, y, z, c=c, cmap=plt.hot())
fig.colorbar(img)
plt.show()
```

CLUSTER

```python
data_name = "eROSITA_cluster_agn_box2hr/cluster_143.dat"
headers2 = ["ID",
            "x [kpc/h]", 
            "y [kpc/h]",
            "z[kpc/h]",
            "Mvir [Msol/h]", 
            "M200m [Msol/h]",
            "M500m [Msol/h]", 
            "M200c [Msol/h]", 
            "M500c [Msol/h]", 
            "R500c [kpc/h]",
            "Mgas_500c [Msol/h]",  
            "Mstar_500c [msol/h]", 
            "T_500c [keV]",
            "Lx_500c [1e44 erg/s]"]
  
table2 = pd.read_csv(data_name, comment='#', sep='\s+', header=None, names=headers2)
table2
```

```python
plt.rcParams["figure.figsize"] = [14.00, 7.00]
#plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

data_table = table2.to_numpy().astype(float)

x = data_table[:,1]
y = data_table[:,2]
z = data_table[:,3]
c = data_table[:,12]

img = ax.scatter(x, y, z, c=c, cmap=plt.hot())
fig.colorbar(img)
plt.show()
```
